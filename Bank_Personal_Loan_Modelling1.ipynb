```python
# Display a frequency distribution for Personal Loan.

fig = plt.figure(figsize=[16,15]);
ax1 = fig.add_subplot(3, 3, 2);
sns.barplot(x=list(loan[0]), y=list(loan [1]), ax=ax1);
ax1.text(0.2, 2800, '{}%'.format(str(round(loan [1] [0]/sum (loan[1])*100,1))), ha='right', va='center', size=13 , fontdict={'weight' : 'bold'})
ax1.text(1.2, 200, '{}%'.format(str(round(loan [1][1]/sum(loan [1])*100,1))), ha='right', va='center',size=13 , fontdict={'weight' : 'bold'})
ax1.set_title('loan', fontdict=axtitle_dict);
ax1.set_xlabel('loan', fontdict=axlab_dict);
ax1.set_ylabel('Count', fontdict=axlab_dict);
ax1.bar_label(ax1.containers[0])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example axtitle_dict (define it if not already)
axtitle_dict = {'fontsize': 12, 'fontweight': 'bold'}

# Create figure with 3 x 3 grid of subplots
fig = plt.figure(figsize=[16, 12])
fig.suptitle('DISTPLOT OF DATA', fontsize=18, fontweight='bold')
fig.subplots_adjust(top=0.92, hspace=0.5, wspace=0.4)

# Ensure Num_cols has at most 9 columns
for i, col in enumerate(Num_cols[:9]):
    ax = fig.add_subplot(3, 3, i+1)
    sns.histplot(bank_data[col], ax=ax, kde=True, color='dodgerblue')

    # Add vertical lines for quantiles, mean, and median
    ax.axvline(bank_data[col].quantile(0.25), color='green', linestyle='--', label='25% Quartile')
    ax.axvline(bank_data[col].mean(), color='red', linestyle='--', label='Mean')
    ax.axvline(bank_data[col].median(), color='purple', linestyle='--', label='Median')
    ax.axvline(bank_data[col].quantile(0.75), color='blue', linestyle='--', label='75% Quartile')

    # Set labels and title
    ax.set_xlabel(col, fontdict=axtitle_dict)
    ax.set_title(f'{col.upper()} | Skewness: {round(bank_data[col].skew(), 3)}', fontdict=axtitle_dict)
    ax.legend(fontsize=10)

plt.show()
```

```python
fig = plt.figure(figsize=[16, 15])
fig.suptitle('Count Plot of Categorical Features', fontsize=18, fontweight='bold')
fig.subplots_adjust(top=0.92, hspace=0.5, wspace=0.4)

for i, col in enumerate(categ_columns[:9]):  # max 9 subplots
    values, counts = np.unique(bank_data[col].dropna(), return_counts=True)

    ax = fig.add_subplot(3, 3, i+1)
    bars = sns.barplot(x=values, y=counts, ax=ax, palette='pastel')

    # Annotate percentage on each bar
    total = sum(counts)
    for j, count in enumerate(counts):
        percentage = round((count / total) * 100, 1)
        ax.text(j, count + max(counts)*0.02, f'{percentage}%', ha='center', fontweight='bold', fontsize=11)

    ax.set_title(f'{col}', fontdict=axtitle_dict)
    ax.set_xlabel(col, fontdict=axlab_dict)
    ax.set_ylabel('Count', fontdict=axlab_dict)

    # Optional: Add numeric bar labels
    for container in ax.containers:
        ax.bar_label(container)

plt.show()
```

```python
# Ensure 'bank_data' is a valid DataFrame
# Select up to 6 numeric columns
Num_cols = bank_data.select_dtypes(include='number').columns[:6]

# Define styling for titles
axtitle_dict = {'fontsize': 12, 'fontweight': 'bold'}

# Create subplots: 2 rows, 3 columns
fig, axs = plt.subplots(2, 3, figsize=[16, 12])
fig.suptitle('DISTPLOT OF DATA', fontsize=18, fontweight='bold')
fig.subplots_adjust(top=0.92, hspace=0.5, wspace=0.4)

# Flatten axs for easier looping if needed (2D -> 1D)
axs = axs.flatten()

for i, col in enumerate(Num_cols):
    ax = axs[i]
    
    # Drop NaNs to avoid errors with seaborn
    data = bank_data[col].dropna()
    
    sns.histplot(data, kde=True, ax=ax, color='dodgerblue')

    # Add statistical lines
    ax.axvline(data.quantile(0.25), color='green', linestyle='--', label='25% Quartile')
    ax.axvline(data.mean(), color='red', linestyle='--', label='Mean')
    ax.axvline(data.median(), color='orange', linestyle='--', label='Median')
    ax.axvline(data.quantile(0.75), color='blue', linestyle='--', label='75% Quartile')

    # Set title and labels
    ax.set_xlabel(col, fontdict=axtitle_dict)
    ax.set_title(f'{col.upper()} | Skewness: {round(data.skew(), 3)}', fontdict=axtitle_dict)
    ax.legend(fontsize=10)

# Hide any unused subplots if Num_cols < 6
for j in range(i+1, len(axs)):
    fig.delaxes(axs[j])

plt.show()
```
